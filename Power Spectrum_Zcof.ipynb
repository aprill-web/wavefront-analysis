```python
df = pd.read_excel("zernike_coefficients.xlsx")  # Reads the first sheet by default
z_list = []  # Initialize an empty list
for i in range(24):
    z_list.append(df.iloc[2:, i+1])  # Append each z_i to the list
print (len(z_list))
```

```python
# function to calculate the power spectrum of the zernike coefficient
def power_spectrum(zernike_coefficient_array):
    averaged_zernike = zernike_coefficient_array - np.mean(zernike_coefficient_array)
    t = averaged_zernike
    sampling_rate = 20
    # performing FFT on Zernike coefficients (1D FFt)
    fft_result = np.fft.fft(t)
    #computing the power spectrum (square of absolute fft)
    ps = np.abs(fft_result)**2
    
    #frequency axis (this is for visuailzation)
    freq = np.fft.fftfreq(len(t), d = 1/sampling_rate)
    return (freq,ps)
```

```python
for i in range(15):
    freqs,power_spec = power_spectrum(z_list[i])
    plt.figure(figsize=(8,6))
    plt.plot(freqs,power_spec)
    plt.grid()
    plt.xlabel('Frequency')
    plt.ylabel('Power Spectrum')
    plt.title(f'Power Spectrum of Zernike Coefficients Z_{i}')
    plt.show()
#     plt.figure(figsize=(10,10))
#     plt.grid()
#     plt.plot(power_spec)
#     plt.show()
```
